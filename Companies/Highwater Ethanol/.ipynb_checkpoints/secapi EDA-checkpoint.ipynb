```python
from sec_api import QueryApi

queryAPI = QueryApi(key)

query = {
  "query": { "query_string": { 
      "query": "ticker:HEOL AND filedAt:{2016-01-01 TO 2023-01-01} AND formType:\"10-Q\"" 
    } },
  "from": "0",
  "size": "10",
  "sort": [{ "filedAt": { "order": "desc" } }]
}

filings = queryAPI.get_filings(query)

print(filings)
```
