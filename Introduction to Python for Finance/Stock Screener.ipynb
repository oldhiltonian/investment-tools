```python
import datetime as dt
import pandas as pd
from pandas_datareader import data as pdr
import yfinance as yf
from tkinter import Tk
from tkinter.filedialog import askopenfilename

yf.pdr_override()
start = dt.datetime(2017, 12, 1)
end = dt.datetime.now()
 
```

```python
root = Tk()
ftypes = [('.xlsm', '.xlsx', '.xls')]
title = 'Title'
dir1 = 'C:\\'
filePath = askopenfilename(filetypes=ftypes, initialdir=dir1, title=title)
file_path = 'RichardStocks.xlsx'

stock_list = pd.read_excel(file_path)
# stock_list = stock_list.head(100)
export_list = []
```

```python
for i in range(400):
    stock = str(stock_list['Symbol'][i])
    RS_rating = stock_list['RS Rating'][i]
    
    
    
    try:
        df = pdr.get_data_yahoo(stock, start, end)
        print('Checking ' + stock + '...')
        
        sma_timeframes = [50, 150, 200]
        for time in sma_timeframes:
            df[f"SMA_{time}"] = round(df.iloc[:, 4].rolling(window=time).mean(), 2)
        
        current_close = df['Adj Close'][-1]
        sma_50_current = df['SMA_50'][-1]
        sma_150_current = df['SMA_150'][-1]
        sma_200_current = df['SMA_200'][-1]
        low_52_week = min(df['Adj Close'][-365:])
        high_52_week = max(df['Adj Close'][-365:])
        
        try:
            moving_average_200_30_days_ago = df['SMA_200'][-30]
        except Exception:
            moving_average_200_30_days_ago = 0
        
        # Condition 1: Current price > 150 SMA and > 200 SMA
        if current_close > sma_150_current and current_close > sma_200_current:
            condition_1 = True
        else:
            condition_1 = False
            
        # Condition 2: 150 SMA > 200 SMA
        if sma_150_current > sma_200_current:
            condition_2 = True
        else:
            condition_2 = False
            
        # Condition 3: 200 SMA trending up for at least 1 month (ideally 4-5 months)
        if sma_200_current > moving_average_200_30_days_ago:
            condition_3 = True
        else:
            condition_3 = False
            
        # Condition 4: 50 SMA > 150 SMA and 50 SMA > 200 SMA
        if sma_50_current > sma_150_current and sma_50_current > sma_200_current:
            condition_4 = True
        else:
            condition_4 = False
            
        # Condition 5: Current price > 50 SMA
        if current_close > sma_50_current:
            condition_5 = True
        else:
            condition_5 = False
            
        # Condition 6: Current price >= 1.3* 52 week low
        if current_close >= 1.3*low_52_week:
            condition_6 = True
        else:
            condition_6 = False
            
        # Condition 7: Current price >= 0.75* 52 week high
        if current_close >= 0.75*high_52_week:
            condition_7 = True
        else:
            condition_7 = False
        
        # Condition 8: IBD RS rating > 70 and the higher the better
        if RS_rating > 70:
            condition_8 = True
        else:
            condition_8 = False
            
        conditions = [condition_1, condition_2, condition_3, condition_4,
                     condition_5, condition_6, condition_7, condition_8]
        
        if all(conditions):
            print('All true')
            export_list.append(stock)
        
    except Exception:
        print('No data on ' + stock)
        
```
