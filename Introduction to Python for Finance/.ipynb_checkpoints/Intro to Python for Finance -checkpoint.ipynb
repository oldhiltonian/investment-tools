Goes over stock ticker data (open, close, high, low) and calculates moving averages

```python
import pandas as pd
import numpy as np
import yfinance as yf
import datetime as dt
from pandas_datareader import data as pdr
```

```python
# Activating the yfinance workaround to access the data
yf.pdr_override()
```

```python
stock = input('Enter a stock ticker symbol: ').upper()
print(f"You entered {stock}")
```

```python
start_year = 2019
start_month = 1
start_day = 1

start = dt.datetime(start_year, start_month, start_day)
end = dt.datetime.now()
```

```python
data = pdr.get_data_yahoo(stock, start, end)
```

```python
data
```

```python
# making a variable-day moving average

average_timeframe = 50 #days
sma_string = f"sma_{average_timeframe}"

data[sma_string] = data.iloc[:, 4].rolling(window=average_timeframe).mean()
```

```python
data

'''Note that the first 50 days will not have a value as there is no possible way to create a 50-day sma in less 
than 50 days

As such, we drop the first 50 rows

'''

data = data.iloc[average_timeframe:]
data
```

```python
'''Lets check is the closing value for each day is above the sma'''

for i in range(len(data)):
    if data.iloc[i]['Close'] < data.iloc[i][sma_string]:
        print('Less than')
    else:
        print('greater than')
```

```python
days_lower = 0
days_higher = 0
for i in data.index:
    if data['Close'][i] < data[sma_string][i]:
        print('Less than')
        days_lower += 1
    else:
        print('Greater than')
        days_higher += 1
        
print(days_higher, days_lower)
```
